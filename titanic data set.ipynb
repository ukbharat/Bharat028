```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
titanic_data.head(10)
```

```python
survived_count = titanic_data['Survived'].value_counts()
plt.bar(survived_count.index, survived_count.values)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count')
plt.show()
```

```python
class_count = titanic_data['Pclass'].value_counts()
plt.pie(class_count.values, labels=class_count.index, autopct='%1.1f%%')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
plt.scatter(titanic_data['Age'], titanic_data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs Age')
plt.show()
```
